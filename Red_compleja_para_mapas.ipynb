```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.Graph()

# Añadir nodos (nodos de ejemplo con coordenadas ficticias)
nodos = {
    1: (32.531, -116.958),  # Universidad
    2: (32.528, -116.956),  # GNC Plaza Alameda Otay
    3: (32.525, -116.952),  # Office Depot Otay
    4: (32.520, -116.950),  # Tecnológico de Tijuana
    5: (32.517, -116.948),  # Carusel
    # Añadir más nodos según sea necesario
}

# Añadir nodos al grafo
for node, coords in nodos.items():
    G.add_node(node, pos=coords)

# Añadir conexiones entre nodos con distancias y pesos
# La estructura es: (nodo1, nodo2, distancia, peso)
aristas = [
    (1, 2, 0.8, 1.5),  # Universidad a GNC Plaza Alameda Otay (tráfico)
    (2, 3, 0.5, 1),    # GNC Plaza Alameda Otay a Office Depot Otay (semáforo)
    (3, 4, 0.6, 0.5),  # Office Depot Otay a Tecnológico de Tijuana (alto)
    (4, 5, 1.0, 0.2),  # Tecnológico de Tijuana a Carusel (sin señales)
    # Añadir más aristas con distancias y pesos
]

for u, v, distancia, peso in aristas:
    G.add_edge(u, v, distance=distancia, weight=peso)

# Visualizar el grafo
pos = nx.get_node_attributes(G, 'pos')
nx.draw(G, pos, with_labels=True, node_color='green', node_size=500, font_size=12, font_weight='bold')
plt.title("Grafo de la Zona de Tijuana")
plt.show()

# Cálculo del camino más corto por distancia (sin considerar los pesos)
camino_distancia = nx.shortest_path(G, source=1, target=5, weight='distance')
print("Camino más corto por distancia:", camino_distancia)

# Cálculo del camino más corto por peso (considerando tráfico, semáforos, etc.)
camino_peso = nx.shortest_path(G, source=1, target=5, weight='weight')
print("Camino más corto por peso:", camino_peso)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.Graph()

# Añadir nodos (nodos de ejemplo con coordenadas ficticias)
nodos = {
    1: (32.531, -116.958),  # Universidad
    2: (32.528, -116.956),  # GNC Plaza Alameda Otay
    3: (32.525, -116.952),  # Office Depot Otay
    4: (32.520, -116.950),  # Tecnológico de Tijuana
    5: (32.517, -116.948),  # Carusel
    6: (32.515, -116.945),  # Plaza Otay
    7: (32.512, -116.940),  # Mercado de Otay
    8: (32.508, -116.930),  # Terminal de Autobuses
    9: (32.503, -116.920),  # Parque de la Amistad
    10: (32.500, -116.915), # Centro de Tijuana
}

# Añadir nodos al grafo
for node, coords in nodos.items():
    G.add_node(node, pos=coords)

# Añadir conexiones entre nodos con distancias, pesos y tipo de conexión
aristas = [
    (1, 2, 0.8, 1.5, 'tráfico pesado'),
    (2, 3, 0.5, 1, 'semáforo'),
    (3, 4, 0.6, 0.5, 'alto'),
    (4, 5, 1.0, 0.2, 'sin señales'),
    (2, 6, 0.7, 1.2, 'calle principal'),
    (6, 7, 0.5, 0.8, 'calle secundaria'),
    (7, 8, 1.5, 1.8, 'puente'),
    (8, 9, 1.2, 1.3, 'túnel'),
    (9, 10, 2.0, 2.5, 'tráfico moderado'),
    (10, 1, 3.0, 3.0, 'calle principal'),
    (3, 5, 0.9, 1.0, 'cruce'),
    (4, 6, 1.1, 1.4, 'calle secundaria'),
]

# Añadir aristas al grafo
for u, v, distancia, peso, tipo in aristas:
    G.add_edge(u, v, distance=distancia, weight=peso, type=tipo)

# Visualizar el grafo
pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=700, font_size=12, font_weight='bold', edge_color='gray')
edge_labels = nx.get_edge_attributes(G, 'type')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Grafo de la Zona de Tijuana (Conexiones Complejas)")
plt.show()

# Cálculo del camino más corto por distancia (sin considerar los pesos)
camino_distancia = nx.shortest_path(G, source=1, target=10, weight='distance')
print("Camino más corto por distancia:", camino_distancia)

# Cálculo del camino más corto por peso (considerando tráfico, semáforos, etc.)
camino_peso = nx.shortest_path(G, source=1, target=10, weight='weight')
print("Camino más corto por peso:", camino_peso)

# Mostrar la información de las aristas
for (u, v, datos) in G.edges(data=True):
    print(f"Arista {u}-{v}: Distancia = {datos['distance']} km, Peso = {datos['weight']}, Tipo = {datos['type']}")
```
